```python
import sys
sys.path.append('..')
from jaewook_header import *
%matplotlib inline
```

```python
datasets=['data1','data2']
spines=['C','L']

PATH={(ds,sp): '../../input/류마티스 데이터/'+ds+'/Anonymized_X_Ray_'+sp+'_spine/'
     for ds in datasets
     for sp in spines}

OUT='../../input/images/'
```

```python
import pydicom
from tqdm import tqdm_notebook as tqdm

for ds in datasets:
    for sp in spines:
        for g in tqdm(glob(PATH[(ds,sp)]+'*.dcm'),desc=ds+'_'+sp):
            p,fe=split(g)
            f,e=splitext(fe)
            dcm=pydicom.dcmread(g)
            img=cv2.normalize(dcm.pixel_array,
                              None,
                              alpha=0,
                              beta=255,
                              norm_type=cv2.NORM_MINMAX,
                              dtype=cv2.CV_8UC1)
            
            if dcm.data_element('PhotometricInterpretation').value[-1]=='1':
                img=255-img
            
            no=re.compile(r'([0-9]*)_').match(f).group(1)
            
            out_path=OUT+ds+'_'+sp+'_'+no.zfill(3)+'.png'
            
            imsave(out_path,img)
```
